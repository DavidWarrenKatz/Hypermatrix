```python
import matplotlib.pyplot as plt
import h5py
import numpy as np
import os
import sys

current_dir = os.getcwd()
parent_dir = os.path.dirname(current_dir)
sys.path.append(parent_dir)
from config_and_print import filtered_list, chrom_file, resolutions, normalization

output_directory = '../../../projects/single_cell_files'

# Ensure resolutions is treated as a tuple or list of strings
if isinstance(resolutions, str):
    resolutions = (resolutions,)

print(f"Resolutions from config: {resolutions}")
resolution_str = resolutions[0]
print(f"Extracted resolution string: {resolution_str}")

def parse_resolution(resolution_str):
    if ':' in resolution_str:
        resolution_value, resolution_label = resolution_str.split(':')
        try:
            resolution = int(resolution_value)
            return resolution, resolution_label
        except ValueError:
            raise ValueError(f"Resolution value should be an integer: '{resolution_value}' in '{resolution_str}'")
    else:
        raise ValueError(f"Invalid resolution format: '{resolution_str}'. Expected format 'value:label', e.g., '1000000:1Mb'.")

resolution, resolution_label = parse_resolution(resolution_str)

def load_tensor(chromosome, resolution_label, output_directory):
    combined_tensor_path = os.path.join(output_directory, f'hic_methy_{resolution_label}_all_cells_tensors/{chromosome}_all_cells_tensor.h5')
    
    if not os.path.exists(combined_tensor_path):
        print(f"Tensor file for {chromosome} does not exist at {combined_tensor_path}")
        return None
    
    with h5py.File(combined_tensor_path, 'r') as hf:
        combined_tensor = hf['Combined_Tensor'][:]
    
    return combined_tensor

# Load the tensor for a specific chromosome
chromosome = 'chr1'
tensor = load_tensor(chromosome, resolution_label, output_directory)

if tensor is not None:
    print(f"Shape of the combined tensor for {chromosome}: {tensor.shape}")
    
    # Plot the heatmap of the first slice
    plt.figure(figsize=(10, 8))
    plt.title(f"Heatmap of the first slice of {chromosome} tensor")
    plt.imshow(tensor[:, :, 0, 50], cmap='viridis', aspect='auto')
    plt.colorbar(label='Value')
    plt.show()

    # Compute and print the norms of matrices for each modality and each sample
    num_samples = tensor.shape[-1]
    num_modalities = tensor.shape[2]

    for sample_idx in range(num_samples):
        print(f"\nNorms for Sample {sample_idx + 1}:")
        for modality_idx in range(num_modalities):
            matrix = tensor[:, :, modality_idx, sample_idx]
            matrix_norm = np.linalg.norm(matrix)
            print(f"Modality {modality_idx}: Norm = {matrix_norm:.4f}")

saved_tensor = tensor
```
