```python
import pandas as pd

# Specify the path to your file
file_path = '../methylHic_example_data/SraRunTable.txt'

# Read the file into a pandas DataFrame
df = pd.read_csv(file_path)

# Display the first 100 rows of the DataFrame
print(df.head(100))

# Get the number of rows
number_of_rows = df.shape[0]

# Get unique assay types
unique_assay_types = df['Assay Type'].unique()

# Get the count of each unique assay type
assay_type_counts = df['Assay Type'].value_counts()

# Display the results
print(f"Number of rows: {number_of_rows}")
print(f"Unique Assay Types: {unique_assay_types}")
print(f"Assay Type Counts:\n{assay_type_counts}")
```
