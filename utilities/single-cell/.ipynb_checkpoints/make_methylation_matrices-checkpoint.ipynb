```python
import pyBigWig
import numpy as np
import h5py
import os
from scipy.stats import ks_2samp
from multiprocessing import Pool
from scipy.sparse import csr_matrix
from scipy.sparse import coo_matrix, csr_matrix, triu, tril

# Chromosome sizes
chromosomes = {
    '1': 249250621, '2': 243199373, '3': 198022430, '4': 191154276, '5': 180915260,
    '6': 171115067, '7': 159138663, '8': 146364022, '9': 141213431, '10': 135534747,
    '11': 135006516, '12': 133851895, '13': 115169878, '14': 107349540, '15': 102531392,
    '16': 90354753, '17': 81195210, '18': 78077248, '19': 59128983, '20': 63025520,
    '21': 48129895, '22': 51304566
}

resolution = 1_000_000
max_distance = int(10_000_000 / resolution) + 1

def pad_matrix_to_shape(matrix, target_shape):
    """Pad the matrix with zeros to match the target shape."""
    if matrix.shape != target_shape:
        padded_matrix = np.zeros(target_shape, dtype=matrix.dtype)
        # Copy the original matrix data into the padded matrix
        padded_matrix[:matrix.shape[0], :matrix.shape[1]] = matrix
        return padded_matrix
    return matrix

def ensure_uniform_shapes(*matrices):
    """Ensure all matrices have the same shape by padding smaller ones. Print details if padding is required."""
    max_rows = max(matrix.shape[0] for matrix in matrices)
    max_cols = max(matrix.shape[1] for matrix in matrices)
    target_shape = (max_rows, max_cols)
    padded_matrices = []
    for matrix in matrices:
        if matrix.shape != target_shape:
            print(f"Padding matrix from shape {matrix.shape} to {target_shape}")
        padded_matrices.append(pad_matrix_to_shape(matrix, target_shape))
    return padded_matrices

def load_hic_data(filepath):
    """Loads Hi-C data from a text file, handles file not found errors."""
    try:
        with open(filepath, 'r') as file:
            data = [line.strip().split() for line in file if len(line.strip().split()) == 3]
    except FileNotFoundError:
        print(f"File not found: {filepath}")
        return []
    return [(int(parts[0]), int(parts[1]), int(parts[2])) for parts in data if all(part.isdigit() for part in parts)]

def create_matrix(data):
    """Creates a symmetric matrix from Hi-C data using sparse format."""
    if not data:
        return None
    max_index = max(max(i, j) for i, j, _ in data)
    rows, cols, vals = zip(*data)
    matrix = coo_matrix((vals, (rows, cols)), shape=(max_index + 1, max_index + 1), dtype=int).tocsr()
    matrix += matrix.T - coo_matrix((matrix.diagonal(), (range(max_index + 1), range(max_index + 1))), shape=(max_index + 1, max_index + 1))
    return matrix

def shift_tensor_to_non_negative(tensor):
    # Find the minimum value in the tensor
    min_value = np.min(tensor)
    
    # If the minimum value is less than 0, shift the tensor
    if min_value < 0:
        shift_value = -min_value  # Calculate the shift value
        shifted_tensor = tensor + shift_value  # Shift all elements
    else:
        shifted_tensor = tensor  # No shift needed if already non-negative

    return shifted_tensor

def emphasize_interactions(matrix, max_distance):
    """"Highlights interactions by adding offsets of contacts. Mimics O/E"""
    emphasized_matrix = csr_matrix(matrix.shape)
    for offset in range(1, max_distance + 1):
        emphasized_matrix += triu(matrix, offset) + tril(matrix, -offset)
    return emphasized_matrix

def extract_vector_from_bigwig(bigwigfile_path, chrom, chrom_length, resolution):
    # Open the BigWig file
    bw = pyBigWig.open(bigwigfile_path)
    
    # Calculate the number of bins needed from the start position to the end of the chromosome
    num_bins = (chrom_length // resolution) + (1 if chrom_length % resolution != 0 else 0)
    
    # Create a vector to store the values
    values_vector = []

    # Fetch values for each bin and append to the vector
    for i in range(num_bins):
        start = i * resolution
        end = start + resolution
        if end > chrom_length:
            end = chrom_length
        
        # Get the mean value in this bin using the BigWig's stats function
        value = bw.stats(chrom, start, end, type='mean')
        
        # The stats function returns a list with one item, or None if no data is available
        if value and value[0] is not None:
            values_vector.append(value[0])
        else:
            values_vector.append(0)  # Assuming missing values should be treated as 0

    # Close the BigWig file
    bw.close()
    
    return values_vector

def load_csr_matrix_from_hdf5(file_path):
    with h5py.File(file_path, 'r') as file:
        data = file['Matrix/data'][:]
        indices = file['Matrix/indices'][:]
        indptr = file['Matrix/indptr'][:]
        shape = file['Matrix'].attrs['shape']  # Assuming shape is stored as an attribute
    return csr_matrix((data, indices, indptr), shape=shape)

def transform_and_symmetrize_alternative(matrix, vector):
    # Ensure matrix and vector are both NumPy arrays
    matrix = np.asarray(matrix)
    vector = np.asarray(vector)

    # Check if vector needs reshaping
    if vector.ndim == 1:
        vector = vector[:, np.newaxis]

    transformed_matrix = np.power(matrix, vector)
    # Take the natural logarithm of the transformed matrix to scale back
    # Handling potential domain issues with log by adding a small number
    safe_transformed_matrix = np.log(transformed_matrix + 1e-10)

    # Symmetrize the matrix by averaging it with its transpose
    symmetric_matrix = (transformed_matrix + transformed_matrix.T) / 2

    return symmetric_matrix

def correlation_matrix(data):
    # Compute the correlation matrix using numpy
    correlation_mat = np.corrcoef(data, rowvar=False)
    return correlation_mat

def process_chrom(bigwigfile_path, chrom, file_prefix, chromosomes, resolution):
    output_dir = f'../../projects/single_cell_files/tensor_1Mb_AB_dir/chr{chrom}/'
    AB_tensor_filename = os.path.join(output_dir, f'{file_prefix}_{chrom}_AB_tensor.h5')

    # Check if the tensor file already exists
    if os.path.exists(AB_tensor_filename):
        print(f"File {AB_tensor_filename} already exists. Skipping.")
        return

    chrom_length = chromosomes[chrom]
    
    base_raw_dir = '../../projects/single_cell_files/hicluster_1Mb_raw_dir/'
    raw_file_path = base_raw_dir + f'chr{chrom}/{file_prefix}_chr{chrom}.txt'
    #print(f"Processing file: {raw_file_path}")
    data = load_hic_data(raw_file_path)
    if not data:
        print(f"No data loaded from {raw_file_path}")
        return

    raw_csr_mat = create_matrix(data)
    
    if raw_csr_mat is None:
        print(f"Failed to create raw matrix from data in {raw_file_path}")
        return   
    
    emphasized_raw_csr_mat = emphasize_interactions(raw_csr_mat, max_distance)
    dense_emphasized_raw_matr = emphasized_raw_csr_mat.toarray()
    hic_symmetric_raw_matrix = dense_emphasized_raw_matr + dense_emphasized_raw_matr.T - np.diag(dense_emphasized_raw_matr.diagonal())
    column_norms = np.linalg.norm(hic_symmetric_raw_matrix, axis=0) 
    column_norms[column_norms == 0] = np.finfo(float).eps
    normalized_hic_raw_matrix = hic_symmetric_raw_matrix / column_norms[np.newaxis, :]
    
    methylation_vector = extract_vector_from_bigwig(bigwigfile_path, chrom, chrom_length, resolution)
    impute_dir = f'/home/dwk681/workspace/multi_omics_hic_clustering/scNOMeHiC_20210127/1Mb/hicluster_1Mb_impute_dir/chr{chrom}/'
    impute_matrix_file = impute_dir + file_prefix + f'_chr{chrom}_pad1_std1_rp0.5_sqrtvc.hdf5'

    # Check if the Hi-C matrix file exists before proceeding
    if not os.path.exists(impute_matrix_file):
        print(f"Hi-C matrix file {impute_matrix_file} does not exist. Skipping.")
        return

    csr_mat = load_csr_matrix_from_hdf5(impute_matrix_file)
    emphasized_csr_mat = emphasize_interactions(csr_mat, max_distance)
    dense_emphasized_matr = emphasized_csr_mat.toarray()  
    hic_symmetric_matrix = dense_emphasized_matr + dense_emphasized_matr.T - np.diag(dense_emphasized_matr.diagonal())
    corr_hic_matrix = correlation_matrix(hic_symmetric_matrix)
    column_norms = np.linalg.norm(corr_hic_matrix, axis=0)
    normalized_hic_matrix = corr_hic_matrix / column_norms[np.newaxis, :]
    
    symmetric_methylation_matrix = transform_and_symmetrize_alternative(hic_symmetric_matrix, methylation_vector)
    corr_methy_matrix = correlation_matrix(symmetric_methylation_matrix)
    column_norms = np.linalg.norm(corr_methy_matrix, axis=0)
    normalized_methy_matrix = corr_methy_matrix / column_norms[np.newaxis, :]
    
    tensor = np.zeros((3, symmetric_methylation_matrix.shape[0], symmetric_methylation_matrix.shape[1]))
    
    normalized_hic_matrix, normalized_methy_matrix, normalized_hic_raw_matrix = ensure_uniform_shapes(
    normalized_hic_matrix, normalized_methy_matrix, normalized_hic_raw_matrix
    )

    # Create a tensor with the padded matrices
    tensor_shape = (3, normalized_hic_matrix.shape[0], normalized_hic_matrix.shape[1])
    tensor = np.zeros(tensor_shape)
    tensor[0, :, :] = normalized_hic_matrix
    tensor[1, :, :] = normalized_methy_matrix
    tensor[2, :, :] = normalized_hic_raw_matrix
    
    tensor = shift_tensor_to_non_negative(tensor)
    
    # Ensure output directory exists before saving
    os.makedirs(output_dir, exist_ok=True)
    AB_tensor_filename = os.path.join(output_dir, f'{file_prefix}_{chrom}_AB_tensor.h5')
    with h5py.File(AB_tensor_filename, 'w') as hf:
        hf.create_dataset('tensor', data=tensor)
    #print(f"Processed and saved {AB_tensor_filename}")

def process_directory(base_path):
    suffix = '.methy_count.b37.bw'
    files = [f for f in os.listdir(base_path) if f.endswith(suffix)]

    for file_name in files:
        full_path = os.path.join(base_path, file_name)
        file_prefix = file_name[:-len(suffix)]
        for chrom in chromosomes:
            process_chrom(full_path, chrom, file_prefix, chromosomes, resolution) 
            
if __name__ == "__main__":
    base_path = '/home/dwk681/workspace/cluster_cells_from_GSE189158_NOMe_HiC/filesFromCluster/bam/methylation/filter_low_qual/'
    process_directory(base_path)
```

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

def plot_heatmap(tensor_file):
    # Load the tensor data from the file
    with h5py.File(tensor_file, 'r') as file:
        tensor = file['tensor'][:]
    
    # Plot each slice of the tensor as a heatmap
    num_slices = tensor.shape[0]
    for i in range(num_slices):
        plt.figure(figsize=(10, 8))
        plt.imshow(tensor[i], aspect='auto', cmap='viridis')
        plt.colorbar()
        plt.title(f'Heatmap of slice {i + 1}')
        plt.xlabel('Columns')
        plt.ylabel('Rows')
        plt.show()
        nonzero_entries = np.count_nonzero(tensor[i])
        matrix_norm = np.linalg.norm(np.nan_to_num(tensor[i]))
        print(f"Number of nonzero entries: {nonzero_entries}")
        print(f"Norm of the matrix: {matrix_norm}")

tensor_file_path = '/home/dwk681/workspace/multi_omics_hic_clustering/scNOMeHiC_20210127/1Mb/tensor_1Mb_AB_dir/chr5/sc17.ACTTGA_5_AB_tensor.h5'
plot_heatmap(tensor_file_path)
```
