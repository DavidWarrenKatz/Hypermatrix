```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
import os

def load_cumulant_tensor(file_path):
    """Load the cumulant tensor from an HDF5 file."""
    with h5py.File(file_path, 'r') as h5file:
        cumulant_tensor = h5file['cumulant_tensor'][:]
    return cumulant_tensor

output_directory = '../../../projects/single_cell_files'
chromosome = 'chr1'  
prefix = 'sc11.CGATGT'  
file_path = os.path.join(output_directory, 'hic_1Mb_cumulant_dir', chromosome, f'{prefix}_{chromosome}_cumulant.h5')
# Load the cumulant tensor
cumulant_tensor = load_cumulant_tensor(file_path)

# Display the shape of the tensor
print(f"Cumulant tensor shape: {cumulant_tensor.shape}")

# Select a slice to visualize
slice_index = 100  # Change this index to view different slices

plt.figure(figsize=(10, 8))
plt.imshow(cumulant_tensor[slice_index], cmap='viridis')
plt.colorbar()
plt.title(f'Cumulant Tensor Slice at Index {slice_index}')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()

# Sum across the third dimension (for example)
aggregated_tensor = np.sum(cumulant_tensor, axis=2)

plt.figure(figsize=(10, 8))
plt.imshow(aggregated_tensor, cmap='viridis')
plt.colorbar()
plt.title('Sum of Cumulant Tensor Across the Third Dimension')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()

# Select a specific element to plot
element_index_1 = 0  # Index for the first dimension
element_index_2 = 0  # Index for the second dimension

plt.figure(figsize=(10, 6))
plt.plot(cumulant_tensor[element_index_1, element_index_2, :])
plt.title(f'1D Slice of Cumulant Tensor at Indices ({element_index_1}, {element_index_2})')
plt.xlabel('Dimension 3')
plt.ylabel('Value')
plt.show()
```

```python
import numpy as np
import h5py
import os

def load_cumulant_tensor(file_path):
    """Load the cumulant tensor from an HDF5 file."""
    with h5py.File(file_path, 'r') as h5file:
        cumulant_tensor = h5file['cumulant_tensor'][:]
    return cumulant_tensor

def normalize_cumulant_tensor(tensor, positive_threshold, negative_threshold):
    """Normalize the cumulant tensor based on the specified thresholds."""
    normalized_tensor = np.zeros_like(tensor)

    # Binarize positive values
    normalized_tensor[tensor > positive_threshold] = positive_threshold
    #normalized_tensor[(tensor > 0) & (tensor <= positive_threshold)] = 0
    
    # Binarize negative values
    normalized_tensor[(tensor < 0) & (np.abs(tensor) > negative_threshold)] = -1
    #normalized_tensor[(tensor < 0) & (np.abs(tensor) <= negative_threshold)] = 0

    return normalized_tensor

output_directory = '../../../projects/single_cell_files'
chromosome = 'chr1'  
prefix = 'sc11.CGATGT'  
file_path = os.path.join(output_directory, 'hic_1Mb_cumulant_dir', chromosome, f'{prefix}_{chromosome}_cumulant.h5')

# Load the cumulant tensor
cumulant_tensor = load_cumulant_tensor(file_path)

# Define thresholds
positive_threshold = 1  # Adjust this threshold as needed
negative_threshold = 0.5  # Adjust this threshold as needed

# Normalize the tensor
normalized_tensor = normalize_cumulant_tensor(cumulant_tensor, positive_threshold, negative_threshold)

# Check the shape and some basic stats after normalization
print(f"Normalized tensor shape: {normalized_tensor.shape}")
print(f"Unique values in normalized tensor: {np.unique(normalized_tensor)}")

# Optionally, save the normalized tensor for further analysis
normalized_file_path = file_path.replace('_cumulant.h5', '_normalized_cumulant.h5')
with h5py.File(normalized_file_path, 'w') as h5file:
    h5file.create_dataset('normalized_cumulant_tensor', data=normalized_tensor)
print(f"Normalized cumulant tensor saved to {normalized_file_path}")

# Select a slice to visualize
slice_index = 200  # Change this index to view different slices

plt.figure(figsize=(10, 8))
plt.imshow(normalized_tensor[slice_index], cmap='viridis')
plt.colorbar()
plt.title(f'Cumulant Tensor Slice at Index {slice_index}')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()

# Sum across the third dimension (for example)
aggregated_tensor = np.sum(normalized_tensor, axis=2)

plt.figure(figsize=(10, 8))
plt.imshow(aggregated_tensor, cmap='viridis')
plt.colorbar()
plt.title('Sum of Cumulant Tensor Across the Third Dimension')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()

```
