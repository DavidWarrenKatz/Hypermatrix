```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def load_tensor(tensor_path):
    with h5py.File(tensor_path, 'r') as hf:
        tensor = hf['Tensor'][:]
    return tensor

def plot_side_by_side_heatmaps(matrix1, matrix2, title1, title2):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))
    
    sns.heatmap(matrix1, cmap='viridis', cbar=True, ax=ax1)
    ax1.set_title(title1)
    
    sns.heatmap(matrix2, cmap='viridis', cbar=True, ax=ax2)
    ax2.set_title(title2)
    
    plt.show()

def print_statistics(matrix, name):
    print(f"Statistics for {name}:")
    print(f"Shape: {matrix.shape}")
    print(f"Min value: {np.nanmin(matrix)}")
    print(f"Max value: {np.nanmax(matrix)}")
    print(f"Frobenius norm: {np.linalg.norm(np.nan_to_num(matrix))}")
    print(f"Number of non-zero elements: {np.count_nonzero(~np.isnan(matrix) & (matrix != 0))}")
    column_norms = np.linalg.norm(np.nan_to_num(matrix), axis=0)
    print(f"Norm of each column: {column_norms}\n")

# Load the tensor
tensor_path = '../../../projects/single_cell_files/hic_methy_1Mb_tensor_singlecell/chr2/sc19.TAGCTT_chr2.h5'
tensor = load_tensor(tensor_path)

# Extract Hi-C and methylation slices
hic_slice = tensor[..., 0]
methy_slice = tensor[..., 1]

# Plot heatmaps side by side
plot_side_by_side_heatmaps(hic_slice, methy_slice, 'Hi-C Matrix Heatmap', 'Methylation Matrix Heatmap')

# Print statistics
print_statistics(hic_slice, 'Hi-C Matrix')
print_statistics(methy_slice, 'Methylation Matrix')
```
