```python
import openpyxl
import h5py
import pandas as pd
import numpy as np
import random
import scipy.io

# Input Excel file and output HDF5 file paths
xlsx_file_path = '/home/dwk681/workspace/GSE143519/mmc1.xlsx'

#1,461 rows and header 
#column 1 is gene name
#column 2 is amino-acid, not sure what this is                NOT USED
#column 3 is T:Sequence Window, not sure what this is         NOT USED
#column 4 Cluster Identity (8 clusters)
#column 5 gene expressison, 0% stress, replicate 1
#column 6 gene expression, 0% stress, replicate 2
#column 7 gene expression, 0% stress, replicate 3
#column 8 gene expression, 40% stress, 30 mins, replicate 1
#column 9 gene expression, 40% stress, 30 mins, replicate 2
#column 10 gene expression, 40% stress, 30 mins, replicate 3
#column 11 gene expression, 40% stress, 6 hours, replicate 1
#colum 12 gene expression, 40% stress, 6 hours, replicate 2
#column 13 gene expression, 40% stress, 6 hours, replicate 3
#column 14 ANOVA always +                                          NOT USED
#column 15 is -log p-value, but p-value of what?                   NOT USED  
```

```python
#n by 2 by 2 hypermatrix

hdf5_file_path = '/home/dwk681/workspace/GSE143519/Workspaces/zscore_0stress_or_40stress_30mins_360mins_noReplicates/'

# Open the Excel file using openpyxl
workbook = openpyxl.load_workbook(xlsx_file_path)

# Select the first sheet (you may need to specify the sheet name if not the first sheet)
sheet = workbook.active 

# Shuffle the rows of the sheet
data_rows = list(sheet.iter_rows(min_row=2, values_only=True))
random.shuffle(data_rows)  # Shuffle the rows

# Create lists for the 8th and 11th columns
column_1 = []
column_3 = []
column_4 = []
column_5 = []
column_6 = []
column_7 = []
column_8 = []
column_9 = []
column_10 = []
column_11 = []
column_12 = []
column_13 = []

head_rows = 5  # You can change this to the number of rows you want to print
print("Head of the shuffled data:")
for i in range(head_rows):
    print(f"Row {i+1}: {data_rows[i]}")

# Iterate through the rows and extract data from the columns
for row in data_rows:
    column_1.append(row[0])
    column_3.append(row[2])
    column_4.append(row[3])
    column_5.append(row[4])   
    column_6.append(row[5])  
    column_7.append(row[6])   
    column_8.append(row[7])
    column_9.append(row[8])   
    column_10.append(row[9])  
    column_11.append(row[10])
    column_12.append(row[11])   
    column_13.append(row[12]) 

gene_to_number = {}
next_gene_number = 1
    
# Create a list to store the assigned numbers based on gene
gene_numbers = []

# Process the data and assign numbers to genes
for gene_name in column_1:
    if gene_name not in gene_to_number:
        gene_to_number[gene_name] = next_gene_number
        next_gene_number += 1
    gene_number = gene_to_number[gene_name]
    gene_numbers.append(gene_number)

# gene_numbers now contains the assigned numbers for each gene
# You can use gene_numbers as your vector with the same length as column 1

# Combine all vectors into a numpy array
all_vectors = np.array([column_5, column_6, column_7, column_8,
                        column_9, column_10, column_11, column_12, column_13])

# Find the minimum value among all vectors
min_value = all_vectors.min()

# Translate all vectors by the absolute value of the minimum value
all_vectors += abs(min_value)

#separate the modified vectors back into individual lists
column_5 = all_vectors[0, :].tolist()
column_6 = all_vectors[1, :].tolist()
column_7 = all_vectors[2, :].tolist()
column_8 = all_vectors[3, :].tolist()
column_9 = all_vectors[4, :].tolist()
column_10 = all_vectors[5, :].tolist()
column_11 = all_vectors[6, :].tolist()
column_12 = all_vectors[7, :].tolist()
column_13 = all_vectors[8, :].tolist()    
    
matrix1 = [column_5, ((np.array(column_6) + np.array(column_7)) / 2).tolist()]
print("Minimum value:", np.nanmin(matrix1))
with h5py.File(hdf5_file_path + "slice_a_1_.mat", 'w') as hf:
    hf.create_dataset('matrix', data=matrix1)
    
matrix2 = [((np.array(column_8) + np.array(column_9) + np.array(column_10))/3).tolist(), ((np.array(column_8) + np.array(column_12) + np.array(column_13))/3).tolist()]
#replicate 2, 40 percent stress, 30 mins and 6 hours
print("Minimum value:", np.nanmin(matrix2))
with h5py.File(hdf5_file_path + "slice_a_2_.mat", 'w') as hf:
    hf.create_dataset('matrix', data=matrix2)
    
column_1 = ['NONE' if value is None or value == '' else value for value in column_1]
column_1_array = np.array(column_1, dtype=object)  # Use dtype=object
scipy.io.savemat(hdf5_file_path + "gene_names.mat", {'gene_names': column_1_array})      
    
with h5py.File(hdf5_file_path + "gene_numers.mat", 'w') as hf:
    hf.create_dataset('matrix', data=gene_numbers)   

cluster_mapping = {
    "Cluster 1": 1,
    "Cluster 2": 2,
    "Cluster 3": 3,
    "Cluster 4": 4,
    "Cluster 5": 5,
    "Cluster 6": 6,
    "Cluster 7": 7,
    "Cluster 8": 8
}

# Replace cluster names with numerical values
column_4_numeric = [cluster_mapping[cluster] for cluster in column_4]    

# Convert column_3_numeric to a NumPy array of integers
column_4_array = np.array(column_4_numeric, dtype=int)

# Save column_3_array as a MATLAB file
scipy.io.savemat(hdf5_file_path + "cluster_names.mat", {'cluster_numbers': column_4_array})
```
